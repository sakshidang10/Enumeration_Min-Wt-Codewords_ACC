```python
n = 5    #length of code
q = 4    # size of finite field
F.<z>=GF(q)
h = 0      # size of I_h
k = 2     # number of rows of G(dimG)

S=SymmetricGroup(n)
P=[];
for i in range(len(S)):
    p=S[i].matrix()
    P.append(p)
print("Total permutation matrices = ", len(P))

import itertools
def invertible_diagonal_matrices(F, n):
# all nonzero field elements
    nonzero = [a for a in F if a != 0]
# product over all possible diagonal entries
    for diag_entries in itertools.product(nonzero, repeat=n):
        yield diagonal_matrix(F, diag_entries)

D = list(invertible_diagonal_matrices(F, n))
print("Total invertible diagonal matrices =", len(D))

M=[];
for p in P:
    for d in D:
        m=p*d;
        M.append(m)
print("Total monomial matrices= ", len(M))
```

```python
C1_A=[];
for A in A_space:
    p1=Ih+A*A.transpose()
#     for i,j in range(h):
    if p1==0: 
#     all(p1[i,j]==0 for i,j in range(h)):
        C1_A.append(A)
# print(C1_A)
print("Matrices with Condition I+AA^t = 0 :",len(C1_A))

C1_B=[];
for B in B_space:
    BBT=B*B.transpose()
    if(BBT.rank()==k-h):
        C1_B.append(B)
# print(C1_B)
print("Matrices with Condition rank(BB^t) = k-h :",len(C1_B))


All_M=[];
for A in C1_A:
    for B in C1_B:
        p2=A*B.transpose()
#         print(p2)
#         if(p2[i,j]==0 for i in range(h) for j in range(k-h)):
        if p2==0:
            All_M.append((A,B))
# print(All_M)
print("Total choices for (A,B) with condition AB^t = 0 :",len(All_M))          
```
